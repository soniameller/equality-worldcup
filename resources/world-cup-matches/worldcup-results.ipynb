## Woman dataset cleanup

Change stage names:

```python
women_worldcup_results = women_worldcup_results.replace(np.nan, '', regex=True)
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group A' if (x == 'group-a') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group B' if (x == 'group-b') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group C' if (x == 'group-c') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group D' if (x == 'group-d') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group E' if (x == 'group-e') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Group F' if (x == 'group-f') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Final' if (x == 'final') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Quarter-finals' if (x == 'quarter-finals') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Semi-finals' if (x == 'semi-finals') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Third place' if (x == 'third-place') else x )
women_worldcup_results['Stage'] = women_worldcup_results['Stage'].apply(lambda x: 'Round of 16' if (x == 'round-16') else x )
```

Separate score to each side:

```python
women_worldcup_results['Score'] = women_worldcup_results['Score'].apply(lambda x: x.replace('-','–'))

women_worldcup_results['Home Team Goals'] = women_worldcup_results['Score'].apply(lambda x: x.split('–')[0])
women_worldcup_results['Away Team Goals'] = women_worldcup_results['Score'].apply(lambda x: x.split('–')[1])

```

```python
women_worldcup_results = women_worldcup_results[['Year','Stage','Home Team Name','Home Team Goals','Away Team Goals','Away Team Name','Win conditions']]
women_worldcup_results.insert(1, 'Gender', 'F')
```

## Probability of winning a match

```python
def probability_of_winning(country, gender_M_or_F):
    
    df = world_cup_results[world_cup_results['Gender'] == gender_M_or_F.upper()]
    
    
    home_team = df[df['Home Team Name'] == country]

    home_loses = len(home_team[(home_team['Home Team Goals']  < home_team['Away Team Goals'])])
    home_wins = len(home_team[(home_team['Home Team Goals']  > home_team['Away Team Goals'])])
    home_draws = len(home_team[(home_team['Home Team Goals']  == home_team['Away Team Goals'])])

    away_team = df[ df['Away Team Name'] == country]
    
    away_loses = len(away_team[(away_team['Home Team Goals']  > away_team['Away Team Goals'])])
    away_wins = len(away_team[(away_team['Home Team Goals']  < away_team['Away Team Goals'])])
    away_draws = len(away_team[(away_team['Home Team Goals']  == away_team['Away Team Goals'])])
    
    loses = home_loses + away_loses
    wins = home_wins + away_wins
    draws = home_draws + away_draws
    total = loses + wins + draws

#     print("loses",home_loses + away_loses)
#     print("Wins",home_wins + away_wins)
#     print("Draws",home_draws + away_draws)
#     print("Total matches played =", total)
    
    if (total == 0):
        return 0
    
    else:
        return wins/total
```

## Add probability of winning to df

```python
world_cup_results.head()
```

```python
new_men = men.copy()
new_women = women.copy()

new_men['home_team_p'] = new_men['Home Team Name'].apply(lambda x: probability_of_winning( x, 'm'))
new_women['home_team_p'] = new_women['Home Team Name'].apply(lambda x: probability_of_winning( x, 'f'))
```

```python
world_cup_p = new_men.append(new_women)
world_cup_p = world_cup_p[['Home Team Name','Gender','home_team_p']]
world_cup_p = world_cup_p.groupby(['Home Team Name','Gender']).first()
world_cup_p.reset_index(inplace=True)
world_cup_p.rename(columns={'Home Team Name': 'Country Name_x', 'home_team_p': 'p'}, inplace = True)
world_cup_p.head(15)
```
